```python
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score

import warnings
warnings.filterwarnings("ignore")
```

```python
previsores = pd.read_csv("entradas-breast.csv")
classe = pd.read_csv("saidas-breast.csv")
```

```python
previsores.head()
```

### Construindo a rede neural utilizando o metodo K-Fold Cross Validation para treinamento e teste.

### k = 10

```python
def criaRede():
    i = 0
    n_HLayers = 2                              #Quantidade de Hidden Layers
    
    classificador = Sequential()
    classificador.add(Dense(units = 16, activation = 'relu', kernel_initializer = 'random_uniform', input_dim = 30))
    
    for i in range(n_HLayers-1):
        classificador.add(Dense(units = 16, activation = 'relu', kernel_initializer = 'random_uniform'))
    
    classificador.add(Dense(units = 1, activation = 'sigmoid'))
    otimizador = keras.optimizers.Adam(decay = 0.0001, clipvalue = 0.5)
    classificador.compile(optimizer = otimizador, loss = 'binary_crossentropy', metrics = ['binary_accuracy'])
    return classificador
```

### O resultado abaixo retorna uma media de todas as K vezes que a rede neural aprendeu com a base de treinamento e testou na base de testes

Note que esse metodo visa criar uma rotação de base de treinamento e testes para que a rede neural possa explorar toda a base de dados

```python
classificador = KerasClassifier(build_fn = criaRede, epochs = 100, batch_size = 10)
resultados = cross_val_score(estimator = classificador, X = previsores, y = classe, cv = 10, scoring = 'accuracy')
```

```python
print("\nResultado: ", resultados.mean(), "%")
```
